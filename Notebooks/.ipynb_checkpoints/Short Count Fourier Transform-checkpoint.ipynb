```python
# Import necessary libraries
import scipy.io.wavfile
import scipy.signal
from IPython.display import Audio
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Wave file, sampling
Read a wave file, get it's sampling rate and length, and display it to play. The wave data is a single dimensional array, where wave[i] is the amplitude of frame i. Each frame represents 1/fs of a second.

```python
filename = 'short_count.wav'

# fs is our sampling rate
fs, wave = scipy.io.wavfile.read(filename)

print ('Data:', wave)
print ('Sampling rate:', fs)
print ('Audio length:', wave.size/fs, 'seconds', )
print ('Lowest amplitude:', min(wave))
print ('Highest amplitude:', max(wave))
print(wave)
Audio(filename)
```

Now plot the wave signal to observe different points in time.

```python
# Normalize using One-Dimensional Interpolation
wave_normed = np.interp(wave,(wave.min(), wave.max()),(-1, 0.5))

# Generate vector for time axis
total_samples = len(wave)
time_seconds = total_samples / fs
time_vector = np.linspace(0, time_seconds, total_samples)

# Plot of original signal
plt.plot(time_vector, wave_normed)

plt.title('train.wav signal plot')
plt.xlabel('time (sec)')
plt.ylabel('amplitude')
```

# 3. Fourier Transforms (FFT)
Given a sound, we apply Fourier Analysis to decompose it into its frequencies, and visualize them.

```python
# Get FFT of original signal
fft_freq = np.fft.fftshift(np.fft.fft(wave_normed))
freq = np.fft.fftshift(np.fft.fftfreq(len(wave_normed), wave_normed))

plt.plot(time_vector, abs(fft_freq))
plt.title('short_count.wav FFT Shifted')
plt.xlabel('Time [sec]')
plt.ylabel('Frequency')
```

```python
print(fft_freq)

# If we play audio with same frequency sample, we get this sound
Audio(fft_freq, rate=fs)
```

# Inverse Fourier Transformation (IFFT)
I choose option C) Mess up the sound with some other change to the spectrum.

```python
# Do an inverse Fourier transform to get amplitude back with Gaussian filter applied

# Apply simple Gaussian filter in the frequency domain
filter = np.exp(- freq**2/6)

# Apply filter to spectrum
filtered_spectrum = fft_freq * filter

ifft_freq = np.fft.ifft(filtered_spectrum)
plt.plot(time_vector, ifft_freq)

plt.title('short_count.wav using IFFT')
plt.xlabel('Time [sec]')
plt.ylabel('Amplitude')

plt.show()

# Play backtransform audio with Gaussian filter applied at same frequency sample
Audio(ifft_freq, rate=fs)
```

# Visualize spectrogram
We'll visualize our original audio file as a spectrogram, with time on the x-axis, frequency on the y-axis, and the intensity of a given frequency at the given time step noted by shading.

```python

def spectrogram(wavfile):
    fs, wave = scipy.io.wavfile.read(wavfile)
    spec = plt.specgram(wave, NFFT=int(fs*0.005), Fs=fs, cmap=plt.cm.gray_r, pad_to=256, noverlap=int(fs*0.0025))
    plt.show()

spectrogram('short_count.wav')
```
